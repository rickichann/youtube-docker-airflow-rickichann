```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('retail-dataset.csv')
df.head()
```

```python
df = df.drop(df.columns[0],axis=1)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'],format='%d-%m-%Y %H:%M')
```
