```python
df = pd.read_csv('retail-dataset.csv')
```

```python
df = df.drop(df.columns[0],axis=1)
```

```python
df['totalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
dfg = df[['Country','StockCode','Description','Quantity','totalPrice']]
```
